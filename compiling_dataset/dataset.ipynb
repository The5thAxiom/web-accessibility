```python
import pandas as pd
```

```python
text = ''
with open('acc.txt') as f:
    text = f.read()
sites = [t.strip() for t in text.split(',')]
sites
```

```python
df = pd.DataFrame()
df['sites'] = sites
for i, row in enumerate(df['site']):
    df.loc[i, 'label'] = 'accessible'
df
```

```python
text = ''
with open('inacc.txt') as f:
    text = f.read()
sites = [t.strip() for t in text.split(',')]
sites
```

```python
df2 = pd.DataFrame()
df2['sites'] = sites
for i, row in enumerate(df2['site']):
    df2.loc[i, 'label'] = 'inaccessible'
df2
```

```python
df3 = pd.concat([df, df2], )
df3
```

```python
df3.to_csv('dataset.csv', index=False)
```
